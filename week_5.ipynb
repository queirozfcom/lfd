```python
import math

def func(N):
    return (0.1 ** 2) * (1 - (float(9) / N))

for (i,_) in enumerate(range(200), start=1):
    print "N={0}, func={1}".format(i,func(i))
```

```python
def get_err(u,v):
    u_dbl = float(u)
    v_dbl = float(v)
    
    in_param = (u_dbl * math.exp(v_dbl)) - ( 2 * v_dbl * math.exp(u_dbl*(-1)) ) 
    
#     print "in param is: {0}".format(in_param)
    
    return in_param ** 2

def get_grad(u,v):
    u_dbl = float(u)
    v_dbl = float(v)
        
    dedu = 2* ( math.exp(v_dbl) + 2*v_dbl*math.exp(u_dbl*(-1)) ) * ( (u_dbl * math.exp(v_dbl)) - ( 2 * v_dbl * math.exp(u_dbl*(-1)) ) )
    dedv = 2* ( u_dbl * math.exp(v_dbl) - 2 * math.exp(u_dbl*(-1)) ) * ( (u_dbl * math.exp(v_dbl)) - ( 2 * v_dbl * math.exp(u_dbl*(-1)) ) )

    return (dedu,dedv)

u = 1
v = 1

for i in xrange(15):
    (du,dv) = get_grad(u,v)
    
    print "du is: {0} and dv is: {1}".format(du,dv)
    
    u = u - float(du) / 10
    
    print "ITERATION {0}: error is: {1}, U={2} and V={3}".format(i,get_err(u,v),u,v)
    
    (du,dv) = get_grad(u,v)
    
    v = v - float(dv) / 10
    
    print "ITERATION {0}: error is: {1}, U={2} and V={3}".format(i,get_err(u,v),u,v)
    
    print ""
    
```
